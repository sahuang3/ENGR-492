```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
```

```python
x = np.arange(0, 4, 0.01)
y = x**4 - 3*x**3 + 2
plt.plot(x,y, 'r', linewidth=1)
```

```python
next_x = 3 # the initial guessing point
gamma = 0.001 # learning rate
precision = 0.0001 # if the change of y is less than the precision, we will stop the process to find the minimum
max_iters = 1000 # max iterations
```

```python
def df(x):
    return 4*x**3 - 9*x**2 # derivative of function
```

```python
xList = []
for _i in range(max_iters):
    current_x = next_x
    next_x = current_x - gamma * df(current_x)
    xList.append(next_x)

    step = next_x - current_x
    if abs(step) <= precision:
        break
```

```python
print("min x", next_x)
```

```python
n = len(xList)
yList = []
for i in range(n):
    yy = xList[i]**4 - 3*xList[i]**3 + 2
    yList.append(yy)
```

```python
x_min = next_x
y_min = x_min**4 -3*x_min**3 + 2
```

```python
plt.plot(x, y, 'r')
plt.plot(x_min, y_min, marker="o", color='red')
plt.plot(xList, yList, marker="+")
```
